```python
# load modules 
import pandas as pd
import geopandas as gpd
from ChelsaHelpFunctions import *
```

### Rainfall

```python
# add data  1988 to 2005 
prc_1988_2005=[]
for year in range(1988,2006): 
    print('year:',year)
    dta_year=get_yearly_climate(year,'Rainfall\Zonal statistics','prc')
    prc_1988_2005.append(dta_year)

prc_1988_2005_df=pd.concat(prc_1988_2005, ignore_index=True, axis=0)
```

```python
# add data  2006 to 2018 
prc_2006_2018=[]
for year in range(2006,2019): 
    print('year:',year)
    dta_year=get_yearly_climate(year,'Rainfall\Zonal statistics','prc')
    prc_2006_2018.append(dta_year)

prc_2006_2019_df=pd.concat(prc_2006_2018, ignore_index=True, axis=0)
```

```python
# save outputs
prc_2006_2019_df.to_csv('prc_2006_2019.csv')
prc_1988_2005_df.to_csv('prc_1988_2005.csv')
```

```python
# join both subsamples
prc_2006_2019_df=pd.read_csv('prc_2006_2019.csv')
prc_1988_2005_df=pd.read_csv('prc_1988_2005.csv')
year_dta=pd.concat([prc_1988_2005_df,prc_2006_2019_df], ignore_index=True, axis=0)
```

### Add rules so that municipalities match

```python
pp_base=muncode_df.copy()
pp_base['munname']=[remove_accent(word) for word in pp_base['munname']]
pp_base['Nombre']=[remove_accent(word) for word in pp_base['Nombre']]
pp_base.replace('PROVIDENCIA','PROVIDENCIA Y SANTA CATALINA',inplace=True)
pp_base.replace('ARCHIPIELAGO DE SAN ANDRES, PROVIDENCIA Y SANTA CATALINA','SAN ANDRES PROVIDENCIA Y SANTA CATALINA',inplace=True)
```

```python
# Create manual rules 

# for non ambigous municipalities 
pp_mun_merge=climate_df.copy()
pp_mun_merge.replace('BARRANCO MINA','BARRANCOMINAS',inplace=True)
pp_mun_merge.replace('CUASPUD','CUASPUD CARLOSAMA',inplace=True)
pp_mun_merge.replace('CURRILLO','CURILLO',inplace=True)
pp_mun_merge.replace('DON MATIAS','DONMATIAS',inplace=True)
pp_mun_merge.replace('CARMEN DE VIBORAL','EL CARMEN DE VIBORAL',inplace=True)
pp_mun_merge.replace('BUGA','GUADALAJARA DE BUGA',inplace=True)
pp_mun_merge.replace('GUICAN','GUICAN DE LA SIERRA',inplace=True)
pp_mun_merge.replace('LOPEZ','LOPEZ DE MICAY',inplace=True)
pp_mun_merge.replace('MO?ITOS','MOÑITOS',inplace=True)
pp_mun_merge.replace('PIENDAMO','PIENDAMO - TUNIA',inplace=True)
pp_mun_merge.replace('LEGUIZAMO','PUERTO LEGUIZAMO',inplace=True)
pp_mun_merge.replace('PURISIMA','PURISIMA DE LA CONCEPCION',inplace=True)
pp_mun_merge.replace('SAN ANDRES SOTAVENTO','SAN ANDRES DE SOTAVENTO',inplace=True)
pp_mun_merge.replace('TUMACO','SAN ANDRES DE TUMACO',inplace=True)
pp_mun_merge.replace('SAN ANTONIO DE TEQUENDAMA','SAN ANTONIO DEL TEQUENDAMA',inplace=True)
pp_mun_merge.replace('SAN CARLOS GUAROA','SAN CARLOS DE GUAROA',inplace=True)
pp_mun_merge.replace('SAN JUAN BETULIA','SAN JUAN DE BETULIA',inplace=True)
pp_mun_merge.replace('SAN JUAN DE RIO SECO','SAN JUAN DE RIOSECO',inplace=True)
pp_mun_merge.replace('SAN PABLO BORBUR','SAN PABLO DE BORBUR',inplace=True)
pp_mun_merge.replace('MARIQUITA','SAN SEBASTIAN DE MARIQUITA',inplace=True)
pp_mun_merge.replace('SAN VICENTE','SAN VICENTE FERRER',inplace=True)
pp_mun_merge.replace('MOMPOS','SANTA CRUZ DE MOMPOX',inplace=True)
pp_mun_merge.replace('SANTAFE DE ANTIOQUIA','SANTA FE DE ANTIOQUIA',inplace=True)
pp_mun_merge.replace('SANTA.ROSA DE OSOS','SANTA ROSA DE OSOS',inplace=True)
pp_mun_merge.replace('SAN ROSA VITERBO','SANTA ROSA DE VITERBO',inplace=True)
pp_mun_merge.replace('SOTARA','SOTARA PAISPAMBA',inplace=True)
pp_mun_merge.replace('UBATE','VILLA DE SAN DIEGO DE UBATE',inplace=True)
pp_mun_merge.replace('VILLA MARIA','VILLAMARIA',inplace=True)
pp_mun_merge.replace('VISTA HERMOSA','VISTAHERMOSA',inplace=True)

# For ambigous municipalities 
new_name_from_dept(pp_mun_merge,'MANAURE BALCON DEL CESAR','MANAURE','CESAR')
new_name_from_dept(pp_mun_merge,'SAN ANDRES DE CUERQUIA','SAN ANDRES','ANTIOQUIA')
new_name_from_dept(pp_mun_merge,'SAN PEDRO DE LOS MILAGROS','SAN PEDRO','ANTIOQUIA')
new_name_from_dept(pp_mun_merge,'EL SANTUARIO','SANTUARIO','ANTIOQUIA')

#pp_mun_merge[(pp_mun_merge['DEPTO']=='CESAR') & (pp_mun_merge['MPIO_CNMBR']=='MANAURE')]['MPIO_CNMBR']='MANAURE BALCON DEL CESAR'
#pp_mun_merge[(pp_mun_merge['DEPTO']=='ANTIOQUIA') & (pp_mun_merge['MPIO_CNMBR']=='SAN ANDRES')]['MPIO_CNMBR']='SAN ANDRES DE CUERQUIA'
#pp_mun_merge[(pp_mun_merge['DEPTO']=='ANTIOQUIA') & (pp_mun_merge['MPIO_CNMBR']=='SAN PEDRO')]['MPIO_CNMBR']='SAN PEDRO DE LOS MILAGROS'

pp_mun_merge['DEPTO']=pp_mun_merge['DEPTO'].str.upper()
#pp_mun_merge['MPIO_CNMBR']=[word.replace(' ','') for word in pp_mun_merge['MPIO_CNMBR']]
```

```python
# merge municipality code on data 
pp_matche_mun=pp_base.merge(pp_mun_merge, how='left', left_on=['munname','Nombre'],right_on=['MPIO_CNMBR','DEPTO'], indicator=True, validate='1:m')
```

```python
# save climate data 
pp_matche_mun.drop(columns=['munname','Nombre','MPIO_CNMBR','DEPTO','_merge'],inplace=True)
pp_matche_mun.to_csv('climate_muncode.csv')
```
